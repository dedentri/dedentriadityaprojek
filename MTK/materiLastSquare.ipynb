```python
import numpy as np
import pandas as pd

# Misalkan Anda memiliki DataFrame dengan kolom 'x' dan 'y'
data = pd.read_excel("D:\SMK\MTK\DATA TJKT 1.xlsx")
data
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
data
```

```python
data = data.rename(columns={'Unnamed: 1': 'x'})
data = data.rename(columns={'Unnamed: 2': 'y'})
data
```

```python
data.drop(0, inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
data.info()
```

```python
data['x'] = data['x'].astype(int)
data['y'] = data['y'].astype(int)
```

```python
data.info()
```

```python

# Membuat kolom baru 'x0' yang berisi satu untuk term independen/intersep
data['x0'] = 1

# Membangun matriks desain X dari kolom 'x0' dan 'x'
X = data[['x0', 'x']].values

# Membuat array target y dari kolom 'y'
y = data['y'].values

# Menerapkan metode kuadrat terkecil
# θ = (X^T X)^-1 X^T y
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

# Sekarang, theta berisi koefisien regresi, theta[0] adalah intersep dan theta[1] adalah kemiringan
intercept, slope = theta

# Menghitung korelasi antara kolom 'x' dan 'y'
correlation = data['x'].corr(data['y'])


# Mencetak hasil regresi
print("Intersep:", intercept)
print("Kemiringan:", slope)
# Mencetak korelasi
print("Korelasi antara kolom 'x' dan 'y':", correlation)
```
