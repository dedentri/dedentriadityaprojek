```python
import pandas as pd
import numpy as np
```

```python
# Misalkan Anda memiliki DataFrame dengan kolom 'x' dan 'y'
data = pd.read_excel("D:\SMK\MTK\DATA TJKT 1.xlsx")
data
```

```python
data = data[['Unnamed: 1','Unnamed: 2']]
data
```

```python
data = data.rename(columns={'Unnamed: 1': 'x'})
data = data.rename(columns={'Unnamed: 2': 'y'})

data.drop(0, inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
data.info()
```

```python
data['x'] = data['x'].astype(int)
data['y'] = data['y'].astype(int)
```

```python
data.info()
```

```python
korelasi = data['x'].corr(data['y'])
print("Korelasi antara x dan y adalah: ", korelasi)
```

```python
# Menghitung koefisien menggunakan metode Least Squares
X = np.vstack([np.ones(len(data)), data['x']]).T
coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(data['y'])

# Membuat garis regresi linier
x_values = np.linspace(data['x'].min(), data['x'].max(), 100)
y_values = coefficients[0] + coefficients[1] * x_values
```

```python
import matplotlib.pyplot as plt
# Visualisasi dengan diagram pencar dan garis regresi linier
plt.figure(figsize=(8, 6))
plt.scatter(data['x'], data['y'], color='blue', label='Data')
plt.plot(x_values, y_values, color='red', label='Regresi Linier')
plt.title('Diagram Pencar Data dengan Garis Regresi Linier')
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')
plt.legend()
plt.grid(True)
plt.show()
```
